```python
import pandas as pd
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,RobustScaler,LabelEncoder
from sklearn.metrics import r2_score,mean_squared_error,root_mean_squared_error
```

```python
df=pd.read_excel("/content/Student_Performance (1).xlsx")
df
```

```python
df.info()
```

```python
x=df.drop('Performance Index',axis=1)
y=df['Performance Index']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

#PREPROCESSING

```python
#Label encoder
label_encoder=LabelEncoder()
```

```python
xtrain['Extracurricular Activities']=label_encoder.fit_transform(xtrain['Extracurricular Activities'])
'''
another method:-
label_encoder.fit(df['Extracurricular Activities'])
df['Extracurricular Activities']=label_encoder.transform(df['Extracurricular Activities'])
'''
```

Model creation

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(xtrain,ytrain)
```

```python
xtest['Extracurricular Activities']=label_encoder.transform(xtest['Extracurricular Activities'])
```

#MODEL EVALUATION
   * Model score
   * Train score

TRAIN SCORE

```python
ypred=lr_model.predict(xtrain)
trainscore=r2_score(ytrain,ypred)
trainscore
```

```python
lr_model.score(xtrain,ytrain)#model score
```

TEST SCORE

```python
ypred1=lr_model.predict(xtest)
testscore=r2_score(ytest,ypred1)
testscore
```

```python
lr_model.score(xtest,ytest)#model score
```

#r2 score conclusion

```python
print(f"Linear Regression train score:{trainscore}")
print(f"Linear Regression test score:{testscore}")
```

#CONCLUSION:
* model is good fit,
* Because: model performance is good on both train and test data

   * bias -> low:-learns or understand the complex patterns

```python
#model creation:not required
# Linear Regression+L2(Ridge)
ridge_regression=Ridge()
```

```python
Rypred=ridge_regression.predict(xtrain)
Rtrainscore=r2_score(ytrain,Rypred)
Rtrainscore
```

```python
Rypred1=ridge_regression.predict(xtest)
Rtestscore=r2_score(ytest,Rypred1)
Rtestscore
```
