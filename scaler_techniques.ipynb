```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
```

```python
df=pd.read_excel("/content/standardscaler1.xlsx")
df
```

```python
x=df.drop('Target',axis=1)
y=df['Target']
```

#without using standardscaler

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=24)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
trainscore=model.score(xtrain,ytrain)
testscore=model.score(xtest,ytest)
print(trainscore)
print(testscore)
```

```python
model.coef_
```

```python
y_pred=model.predict(xtrain)
```

```python
y_pred1=model.predict(xtest)
```

```python
r2_score(ytrain,y_pred)
```

```python
r2_score(ytest,y_pred1)
```

#with using standardscaler

```python
scaler=StandardScaler()
```

```python
xtrain_scaled=scaler.fit_transform(xtrain)
xtest_scaled=scaler.transform(xtest)
```

```python
model.fit(xtrain_scaled,ytrain)
```

```python
testscore=model.score(xtest_scaled,ytest)
trainscore=model.score(xtrain_scaled,ytrain)
print(testscore)
print(trainscore)
```

```python
model.coef_
```
