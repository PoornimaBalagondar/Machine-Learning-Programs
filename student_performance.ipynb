```python
df=pd.read_excel("/content/Student_Performance (1).xlsx")
df
```

```python
num_df=df.select_dtypes(include="number")
num_df
```

```python
x=num_df.drop('Performance Index',axis=1)#num_df[['H','P']]
y=num_df['Performance Index']
```
