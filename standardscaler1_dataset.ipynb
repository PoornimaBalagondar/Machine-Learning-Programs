```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_excel("/content/standard_scaler_dataset.xlsx")
df
```

```python
x=df.drop('Experience',axis=1)
y=df['Experience']
```

#without using standardscaler

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=24)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
trainscore=model.score(xtrain,ytrain)
testscore=model.score(xtest,ytest)
print(trainscore)
print(testscore)
```

```python
y_pred=model.predict(xtrain)
```

```python
y_pred1=model.predict(xtest)
```

```python
r2_score(ytrain,y_pred)
```

```python
r2_score(ytest,y_pred1)
```

#with using standardscaler

```python
scaler=StandardScaler()
```

```python
xtrain_scaled=scaler.fit_transform(xtrain)
xtest_scaled=scaler.transform(xtest)
```

```python
model.fit(xtrain_scaled,ytrain)
```

```python
testscore=model.score(xtest_scaled,ytest)
testscore
```

```python
trainscore=model.score(xtrain_scaled,ytrain)
trainscore
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,RobustScaler,OrdinalEncoder,StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
import seaborn as sns
```

```python
num_col=xtrain.select_dtypes(include='number').columns
```

```python
preprocessing=ColumnTransformer(
    transformers=[
        ('scaler',StandardScaler(),num_col),
    ]
)
```

```python
decisiontreepipeline=Pipeline(
    steps=[
        ('preprocessing',preprocessing),
        ('model',DecisionTreeRegressor())
    ]
)
decisiontreepipeline.fit(xtrain,ytrain)
```

```python
print("Train_score:",decisiontreepipeline.score(xtrain,ytrain))
print("Test_score:",decisiontreepipeline.score(xtest,ytest))
```
