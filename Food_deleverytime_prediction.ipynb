#Without using standard scaler

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

* distance_km:For every additional 1 km, delivery time increases by about 3 minutes.
* preparation time:For each 1 extra minute of food preparation, total delivery time increases by about 1 minute.
* courier experience:For each extra year of experience, delivery time decreases by about 0.6 minutes.

```python
xtest
```

```python
ytest
```

```python
trainscore=model.score(xtrain,ytrain)#70% model predicted correct values on trained data
testscore=model.score(xtest,ytest)#73% model predicted correct values on test data
print(trainscore)
print(testscore)
```

```python
y_pred=model.predict(xtrain)
```

```python
y_pred1=model.predict(xtest)
```

```python
r2_score(ytrain,y_pred)
```

```python
r2_score(ytest,y_pred1)
```

```python
target_mean=ytest.mean()
target_mean
```

```python
mse=mean_squared_error(ytest,y_pred1)
```

```python
rmse=root_mean_squared_error(ytest,y_pred1)
np.sqrt(mse)
```

```python
model_evaluation=rmse/target_mean
model_evaluation
```
