#Label Encoder,Onehot Encoder

```python
import pandas as pd
```

```python
df=pd.DataFrame({'color':['red','green','blue','red','blue'],'level':['High','Low','High','Medium','Low'],'city':['Bangalore','Haveri','Bangalore','Hubli','Hubli'],'purchased':['Yes','Yes','No','Yes','No']})
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
label_encoder.fit_transform(df['purchased'])
label_encoder.transform(['Yes'])
```

#how to handle unknown values

```python
'''
ordinal=OrdinalEncoder(categories=[['red','green','blue']],handle_unknown='use_encoded_value')
values=ordinal.fit_transform(df[['color']])
ordinal.transform([['orange']])
TypeError: unknown_value should be an integer or np.nan when handle_unknown is 'use_encoded_value', got None.
'''
```

```python
'''
ordinal=OrdinalEncoder(categories=[['red','green','blue']],handle_unknown='use_encoded_value',unknown_value=0)
values=ordinal.fit_transform(df[['color']])
ordinal.transform([['orange']])
ValueError: The used value for unknown_value 0 is one of the values already used for encoding the seen categories.
'''
```

```python
ordinal=OrdinalEncoder(categories=[['red','green','blue']],handle_unknown='use_encoded_value',unknown_value=-1)
values=ordinal.fit_transform(df[['color']])
ordinal.transform([['orange']])
```
